```python
! pip3 install requests
! pip3 install beautifulsoup4
import requests as rq
from bs4 import BeautifulSoup
import re
```

```python
cnn_url = 'http://lite.cnn.com/en'
cnn_page = rq.get(cnn_url)
# print(cnn_page.status_code) # проверка линка
# print(cnn_page.text) # сырой текст
cnn_page.encoding = 'utf-8'
cnn_page_soup = BeautifulSoup(cnn_page.text, 'html.parser') 
# print(cnn_page_soup.prettify()) # форматированный текст
cnn_count = 0
with open('cnn_articles.txt', 'w', encoding = 'utf-8') as cnn_articles: # создаем новый док
    cnn_articles.write('CNN Articles')
for cnn_header_tag in cnn_page_soup.find_all('a'):     
    if 'article' in cnn_header_tag['href']: # почему []
        cnn_count += 1
        cnn_article_num = str(cnn_count)
        cnn_article_header = cnn_header_tag.text # сохраняем заголовок
        cnn_article_url = f'{cnn_url[:-3]}{cnn_header_tag["href"]}' # что такое f'', почему {}
        cnn_article_page = (rq.get(cnn_article_url)).text # ответ сервера на запрос + исходный код страницы
        cnn_article_page_soup = BeautifulSoup(cnn_article_page, 'html.parser') # 'суп'
        cnn_article_text = ' '.join(cnn_article_tag.text for cnn_article_tag in cnn_article_page_soup.find_all('p'))
        # достаем все тэги for TAG in SOUP.find_all('p')) и из каждого тэга делаем текст TAG.text
        # и все эти тексты соединяем в один по переносу строки '\n'.join
        cnn_article_full = cnn_article_num + ' ' + cnn_article_header + '\n' + cnn_article_url + '\n' + cnn_article_text + '\n'
        print(cnn_article_full)
        with open('cnn_articles.txt', 'a', encoding = 'utf-8') as cnn_articles:
            cnn_articles.write(cnn_article_full)
            cnn_articles.write('\n')
with open('cnn_articles.txt', 'r', encoding = 'utf-8') as cnn_articles:
    print(cnn_articles.read())
    
```
