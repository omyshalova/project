```python
url_ling_main = 'https://www.linguee.ru/%D1%80%D1%83%D1%81%D1%81%D0%BA%D0%B8%D0%B9-%D0%B0%D0%BD%D0%B3%D0%BB%D0%B8%D0%B9%D1%81%D0%BA%D0%B8%D0%B9/search?source=auto&query='
url_ling_word = ['dynamic' , 'zenith' , 'aroma' , 'fragrant' , 'fable' , 'calamity' , 'gaudy' , 'stray' , 'grim' , 'serene' , 'innocuous' , 'indifferent' , 'reckless' , 'thrifty' , 'infinite' , 'barren' , 'futile' , 'ruthless' , 'impartial']
count = 0
for item_ling in url_ling_word:
    url_ling_list = url_ling_main + item_ling
    page_ling = rq.get(url_ling_list)
    page_ling.encoding = 'utf-8'
    soup_ling = BeautifulSoup(page_ling.text, 'html.parser') 
    count += 1
    print(count, url_ling_list, page_ling.status_code)
    print('')
    for word_ling in soup_ling.find_all('span', class_='tag_lemma'):
        print(word_ling.text)
        print('')
    for word_ling in soup_ling.find_all('h3', class_='translation_desc'):
        print(word_ling.text)
        print('')
    
```
