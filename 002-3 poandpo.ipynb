```python
! pip3 install requests
! pip3 install beautifulsoup4
import requests as rq
from bs4 import BeautifulSoup
import re
```

```python
poa_url = 'https://lite.poandpo.com/'
poa_page = rq.get(poa_url)
poa_page.encoding = 'utf-8'
poa_page_soup = BeautifulSoup(poa_page.text, 'html.parser') 
poa_count = 0

# проверка линка, сырого и форматированного текста

# print(poa_page.status_code)
# print(poa_page.text)
# print(poa_page_soup.prettify())

with open('poa_articles.txt', 'w', encoding = 'utf-8') as poa_articles: # создаем новый док
    poa_articles.write('poa Articles')

for poa_navigation_tag in poa_page_soup.find_all('a', class_='paddingdesno'):
    poa_count += 1
    poa_navigation_num = str(poa_count) + ' - '
    poa_navigation_str = str(poa_navigation_tag)
    poa_navigation_url = ((str(poa_navigation_str)[30:]).split('"'))[0]
    poa_navigation_page = rq.get(poa_navigation_url)
    poa_navigation_page.encoding = 'utf-8'
    poa_navigation_page_soup = BeautifulSoup(poa_navigation_page.text, 'html.parser') 
    poa_count_1 = 0
    full_poa_navigation = '\n' + poa_navigation_num + poa_navigation_url + '\n'
    with open('poa_articles.txt', 'a', encoding = 'utf-8') as poa_articles:
        poa_articles.write(full_poa_navigation) 
    for poa_article_tag in poa_navigation_page_soup.find_all('h1', itemprop='name headline'): 
        poa_count_1 += 1
        poa_article_num = str(poa_count_1) + ' - '
        poa_article_header = poa_article_tag.text
        poa_article_str = str(poa_article_tag)
        poa_article_url = ((str(poa_article_str)[38:]).split('"'))[0]
        poa_article_title = '\n' + poa_article_num + poa_article_header + '\n' + poa_article_url +'\n'
        poa_article_page = rq.get(poa_article_url)
        poa_article_page.encoding = 'utf-8'
        poa_article_page_soup = BeautifulSoup(poa_article_page.text, 'html.parser') 
        for poa_article_tag in poa_article_page_soup.find_all('article'):
            poa_article_tag = (poa_article_tag.text).replace('Article continues below\n\n', '')
            poa_article = ' '.join(((poa_article_tag.replace('\r', '')).split('\n'))[:1] + ((poa_article_tag.replace('\r', '')).split('\n'))[2:])
        full_poa_article = poa_article_title + poa_article + '\n' 
        with open('poa_articles.txt', 'a', encoding = 'utf-8') as poa_articles:
            poa_articles.write(full_poa_article)    
    
with open('poa_articles.txt', 'r', encoding = 'utf-8') as poa_articles:
    print(poa_articles.read())
```
