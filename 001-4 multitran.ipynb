```python
import requests as rq

from bs4 import BeautifulSoup
```

```python
import re
```

```python
count = 0
url_mtran_main_1 = 'https://www.multitran.com/m.exe?l1=1&l2=2&s='
url_mtran_main_2 = '&langlist=2'
url_mtran_word = ['affluent' , 'prosperous', 'thriving']
for item in url_mtran_word:
    count += 1
    url_mtran_link = url_mtran_main_1 + item + url_mtran_main_2
    page_mtran = rq.get(url_mtran_link)
    page_mtran.encoding = 'utf-8'
    soup_mtran = BeautifulSoup(page_mtran.text, 'html.parser') 
    word_mtran_all = str()
    print(count, 'English word: ', url_mtran_link)
    print(item)    
    print('')
    for word_mtran in soup_mtran.find_all('td', class_='trans'):
        word_mtran = word_mtran.text
        word_mtran = word_mtran.lower()
        word_mtran = re.sub(r'\(.*?\)', '', word_mtran)     # убираем все, что в скобках
        word_mtran = re.sub(r'[a-z]*?', '', word_mtran)     # убираем все, что написано латиницей
        word_mtran = re.sub(r'\)*?\(*?,*?', '', word_mtran) # убираем отдельные скобки
        word_mtran = re.sub(r'\s\s+?', '', word_mtran)      # убираем больше двух пробелов
        word_mtran = re.sub(r'\s;', ';', word_mtran)        # убираем пробел перед ;
        word_mtran = re.sub(r';\s', ';', word_mtran)        # убираем пробел после ;
        word_mtran = word_mtran.strip()                     # убираем пробелы в начале и конце строки
        word_mtran = re.sub(r'$', ';', word_mtran)          # в конце строки ставим ;
        word_mtran_all = word_mtran_all + word_mtran
    print('Translation: ')
    word_mtran_all = word_mtran_all.split(';')
    word_mtran_all = set(word_mtran_all)
    word_mtran_all = list(word_mtran_all)
    word_mtran_all = sorted(word_mtran_all)
    print(word_mtran_all)
#     dic_ruseng_mtran = dict.fromkeys(word_mtran_all,item)
#     print('rus:eng: ')
#     print(dic_ruseng_mtran)
#     dic_engrus_mtran = {}
#     dic_engrus_mtran = {item: word_mtran_all}
#     print('eng:rus: ')
#     print(dic_engrus_mtran)
#     print('')
```
