```python
url_mwd = 'https://www.merriam-webster.com/dictionary/congenial'
```

```python
page_mwd = rq.get(url_mwd)
```

```python
page_mwd.encoding = 'utf-8'
```

```python
page_mwd.status_code
```

```python
page_mwd.text
```

```python
soup_mwd = BeautifulSoup(page_mwd.text, 'html.parser') 
```

```python
print(soup_mwd.prettify())
```

```python
for main_word_mwd in soup_mwd.find_all('h1', class_='hword'):
    print(main_word_mwd.text)
```

```python
for definition_mwd in soup_mwd.find_all('span', class_='dtText'):
    print(definition_mwd.text)
```

```python
for syn_mwd in soup_mwd.find_all('ul', class_="mw-list"):
    print(syn_mwd.text)
```
