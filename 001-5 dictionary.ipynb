```python
soup_dic = BeautifulSoup(page_dic.text, 'html.parser') 
```

```python
print(soup_dic.prettify())
```

```python
for main_word_dic in soup_dic.find_all('h1', class_='css-2m2rhw e1wg9v5m3'):
    print(main_word_dic.text)
```
